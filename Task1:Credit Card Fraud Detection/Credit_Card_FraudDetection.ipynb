```python
import pandas as pd
from collections import Counter
from imblearn.over_sampling import SMOTE

# Loading the creditcard.csv file into a Pandas dataFrame
df = pd.read_csv('creditcard.csv')

# Checking for NaN values in the target variable 'Class'
print(df['Class'].isnull().sum())

# Dropping the rows with NaN values from the target variable 'Class'
df.dropna(subset=['Class'], inplace=True)

# Splitting the creditcard_dataFrame into features (X) and target variable (Y)
X = df.drop('Class', axis=1)
Y = df['Class']

# Applying SMOTE algorithm for handling the class imbalance of the dataset
s = SMOTE(random_state=42)
X_new, Y_new = s.fit_resample(X, Y)

print('Original creditcard_dataset shape is : %s' % Counter (Y))
print('Resampled creditcard_dataset shape is : %s' % Counter(Y_new))
```

```python
#printing the shape or dimension values of the to subsets
print(X_new.shape)
print(Y_new.shape)
```

```python
#scaling the input features
scaler = preprocessing.MinMaxScaler()
names = X.columns
d = scaler.fit_transform(X_new)
scaled_data = pd.DataFrame(d, columns=names)
scaled_data.head()
```

```python
#printing the descriptive statistics for the scaled_data
scaled_data.describe()
```

**Training and Testing creditcard_data**

```python
#splitting the creditcard_dataset into testing and training sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_new, Y_new, test_size = 0.2, random_state = 2)
```

*LOGISTIC REGRESSION MODEL*

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Creating and training the Logistic Regression model
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, Y_train)

# Making predictions on the test set and calculating the accuracy
y_pred = lr.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
accuracy_text = "{:.4f}".format(accuracy)

# Printing the accuracy value
print(f"Accuracy of Logistic Regression model is: {accuracy_text}")

```

*NAIVE BAYES MODEL*

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Creating and training the Naive Bayes model
nb = GaussianNB()
nb.fit(X_train, Y_train)

# Making predictions on the test set and calculating the accuracy
y_pred = nb.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
accuracy_text = "{:.4f}".format(accuracy)

# Printing the accuracy value
print(f"Accuracy of Naive Bayes model is: {accuracy_text}")
```

*DECISION TREE MODEL*

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Creating and training the Decision Tree model
dt = DecisionTreeClassifier(criterion="entropy", random_state=0)
dt.fit(X_train, Y_train)

# Making the predictions on the test set and calculating the accuracy
y_pred = dt.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
accuracy_text = "{:.4f}".format(accuracy)

# Printing the accuracy value
print(f"Accuracy of Decision Tree model is: {accuracy_text}")
```

*LINEAR DISCRIMINANT ANALYSIS (LDA) MODEL*


```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score

# Creating and training the Linear Discriminant Analysis model
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, Y_train)

# Making predictions on the test set and calculating the accuracy
y_pred = lda.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
accuracy_text = "{:.4f}".format(accuracy)

# Printing the accuracy value
print(f"Accuracy of Linear Discriminant Analysis model is: {accuracy_text}")
```
